```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
import numpy as np
```

```python
h = ['ANO', 'UF', 'CIDADE', 'ZONA', 'SECAO', 'SEXO', 'FAIXA_ETARIA', 'GRAU_DE_INSTRUCAO', 'QNT_ELEITORES']
```

```python
df = pd.read_csv('perfil_eleitorado_ATUAL.txt',delimiter=';',encoding = 'latin-1', names=h)
```

```python
groupedByAge = df.groupby(['FAIXA_ETARIA', 'SEXO'])['FAIXA_ETARIA'].count().unstack(1)
groupedByUF = df.groupby(['UF', 'SEXO'])['UF'].count().unstack(1)
groupedByDeg = df.groupby(['GRAU_DE_INSTRUCAO', 'SEXO'])['GRAU_DE_INSTRUCAO'].count().unstack(1)
```

```python
groupedByAge.plot.bar(title='Perfil do eleitorado por faixa etaria', figsize=(16, 8), rot=0)
```

```python
groupedByUF.plot.bar(title='Perfil do eleitorado por UF', figsize=(16, 8), rot=0)
```

```python
#Para o caso do groupby do  UF, acho que pode ser mais interessante usar 
#os graficos com stacked em True
groupedByUF.plot.bar(title='Perfil do eleitorado por UF', figsize=(16, 8), rot=0, stacked=True)
```

```python
groupedByDeg.plot.bar(title='Perfil do eleitorado por grau de escolaridade', figsize=(16, 8))
```
